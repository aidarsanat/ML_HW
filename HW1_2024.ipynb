## Task 1.1 (1 point)

Let $\boldsymbol A \in\mathbb R^{m\times n}$, $\boldsymbol B \in\mathbb R^{n\times m}$. Prove that $\mathrm{tr}(\boldsymbol{AB}) = \mathrm{tr}(\boldsymbol{BA})$. Using this property, calculate $\mathrm{tr}(\boldsymbol{uv}^\mathsf{T})$ if $\boldsymbol u, \boldsymbol v \in\mathbb R^n$, $\boldsymbol u \perp \boldsymbol v$.

### YOUR SOLUTION HERE

1. Мы знаем, что след произведения двух матриц можно записать как сумму произведений их элементов. Для любых матриц $ A \in \mathbb{R}^{m \times n} $ и $ B \in \mathbb{R}^{n \times m} $:
   $$
   \mathrm{tr}(AB) = \sum_{i=1}^{m} (AB)_{ii}
   $$
   где $ (AB)_{ii} $ — это элементы на диагонали матрицы $ AB $.

2. Элемент $ (AB)_{ii} $ — это сумма произведений строк матрицы $ A $ и столбцов матрицы $ B $, что можно переписать как:
   $$
   \mathrm{tr}(AB) = \sum_{i=1}^{m} \sum_{j=1}^{n} A_{ij} B_{ji}
   $$

3. Аналогично:
   $$
   \mathrm{tr}(BA) = \sum_{i=1}^{n} \sum_{j=1}^{m} B_{ij} A_{ji}
   $$

4. Поскольку в обеих формулах элементы находятся в одинаковом порядке, следует, что:
   $$
   \mathrm{tr}(AB) = \mathrm{tr}(BA)
   $$

5. Теперь, для матриц $ u \in \mathbb{R}^n $ и $ v \in \mathbb{R}^n $, где $ u \perp v $ (ортогональны, то есть $ u^T v = 0 $):
   $$
   \mathrm{tr}(uv^T) = \mathrm{tr}(v^T u) = u^T v = 0
   $$
   Таким образом, $ \mathrm{tr}(uv^T) = 0 $.

## Task 1.2 (1 point)

Let $\boldsymbol A \in \mathbb R^{m\times n}$. Prove that $N(\boldsymbol A) = N(\boldsymbol A^{\mathsf T}\boldsymbol A)$, i.e., matrices $\boldsymbol A$ and $\boldsymbol A^{\mathsf T} \boldsymbol A$ have same nullspaces.

### YOUR SOLUTION HERE

1. Ядро матрицы $ A $, обозначенное как $ N(A) $, состоит из всех векторов $ x $, таких что:
   $$
   A x = 0
   $$

2. Рассмотрим ядро матрицы $ A^T A $. Для любого вектора $ x \in N(A^T A) $:
   $$
   A^T A x = 0
   $$
   Домножим на $ x^T $ слева:
   $$
   x^T A^T A x = 0
   $$
   Поскольку $ x^T A^T = (A x)^T $, имеем:
   $$
   (A x)^T (A x) = \|A x\|^2 = 0
   $$

3. Следовательно, $ \|A x\| = 0 $, а это возможно только если $ A x = 0 $. Таким образом, $ x \in N(A) $.

4. Покажем обратное. Если $ x \in N(A) $, то $ A x = 0 $, и тогда:
   $$
   A^T A x = A^T 0 = 0
   $$
   Следовательно, $ x \in N(A^T A) $.

5. Таким образом, ядра матриц $ A $ и $ A^T A $ совпадают:
   $$
   N(A) = N(A^T A)
   $$

## Task 1.3 (0.5 points)

Show that similar matrices have equal determinants: $\det \boldsymbol A = \det \boldsymbol B$ if $\boldsymbol A \sim \boldsymbol B$.

### YOUR SOLUTION HERE

1. Пусть матрицы $ A $ и $ B $ подобны, то есть существует обратимая матрица $ P $, такая что:
   $$
   B = P^{-1} A P
   $$

2. Определитель произведения матриц удовлетворяет свойству:
   $$
   \det(P^{-1} A P) = \det(P^{-1}) \det(A) \det(P)
   $$

3. Поскольку $ \det(P^{-1}) = \frac{1}{\det(P)} $, имеем:
   $$
   \det(B) = \frac{1}{\det(P)} \det(A) \det(P) = \det(A)
   $$

4. Следовательно, у подобных матриц одинаковые определители:
   $$
   \det(A) = \det(B)
   $$

## Task 1.4 (Sherman—Morrison formula, 1.5 points)

Let $\boldsymbol A \in \mathbb R^{n\times n}$ be an invertible matrix, $\boldsymbol u, \boldsymbol v \in \mathbb R^n$ and $\boldsymbol v^\mathsf{T}\boldsymbol A^{-1} \boldsymbol u \ne -1$ . Prove that

$$
    (\boldsymbol A + \boldsymbol u \boldsymbol v^\mathsf{T})^{-1} =
    \boldsymbol A^{-1} - \frac{\boldsymbol A^{-1} \boldsymbol u \boldsymbol v^\mathsf{T} \boldsymbol A^{-1}}{1 + \boldsymbol v^\mathsf{T}\boldsymbol A^{-1} \boldsymbol u }.
$$

### YOUR SOLUTION HERE

1. Для доказательства этой формулы используем предположение, что матрица $ A $ обратима и обратимая матрица $ A + uv^T $ также существует при выполнении условия $ v^T A^{-1} u \neq -1 $.

2. Мы хотим найти $$ (A + uv^T)^{-1}  $$. Начнем с того, что:
   $$
   (A + uv^T) (A^{-1} - \frac{A^{-1} uv^T A^{-1}}{1 + v^T A^{-1} u})
   $$

3. Раскроем скобки:
   $$
   = A A^{-1} - A \frac{A^{-1} uv^T A^{-1}}{1 + v^T A^{-1} u} + uv^T A^{-1} - uv^T A^{-1} \frac{A^{-1} uv^T A^{-1}}{1 + v^T A^{-1} u}
   $$

4. Упрощаем:
   $$
   = I - \frac{uv^T A^{-1}}{1 + v^T A^{-1} u} + uv^T A^{-1} - \frac{uv^T A^{-1} uv^T A^{-1}}{1 + v^T A^{-1} u}
   $$

5. Заметим, что два средних слагаемых сокращаются, и результат будет равен единичной матрице $ I $. Следовательно, доказано:
   $$
   (A + uv^T)^{-1} = A^{-1} - \frac{A^{-1} uv^T A^{-1}}{1 + v^T A^{-1} u}
   $$

### Split the datasert into train and test parts

```python
from sklearn.model_selection import train_test_split

# Разделение данных на тренировочные и тестовые
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

print(X_train.shape, X_test.shape)
```

### Train several models with different hyperparameters (take 1<<k<<20 and different distance metrics ( p=1, p=2, p=+∞))

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Значения k (количество соседей) и метрик расстояния p
k_values = list(range(1, 20))
distance_metrics = [1, 2, np.inf]  # p = 1 (Манхэттен), p = 2 (Евклид), p = ∞ (Чебышев)

train_accuracies = []
test_accuracies = []

# Цикл по метрикам расстояния и значению k
for p in distance_metrics:
    for k in k_values:
        # Инициализация и обучение модели k-NN
        knn = KNeighborsClassifier(n_neighbors=k, p=p)
        knn.fit(X_train, Y_train)

        # Оценка на тренировочной выборке
        train_accuracy = accuracy_score(Y_train, knn.predict(X_train))
        train_accuracies.append((k, p, train_accuracy))

        # Оценка на тестовой выборке
        test_accuracy = accuracy_score(Y_test, knn.predict(X_test))
        test_accuracies.append((k, p, test_accuracy))

        # Печать результата
        print(f'k={k}, p={p}, Train accuracy={train_accuracy:.4f}, Test accuracy={test_accuracy:.4f}')
```

## Task 1.6 (programming, 1.5 points)

Compare the performance of different algorithms for calculation matrix product 

$$
       \boldsymbol C = \boldsymbol {AB}, \quad \boldsymbol A \in \mathbb R^{m\times n}, \quad \boldsymbol B \in \mathbb R^{n\times p}.
$$

Try the following methods:

* nested pythonic loop implementing the formula $C_{ik} = \sum\limits_{j=1}^n A_{ij}B_{jk}$
* replace the inner `for` loop by `np.dot` for calculating $\boldsymbol a_i^\mathsf{T}\boldsymbol b_k$
* calculate $\sum\limits_{j=1}^n \boldsymbol a_j \boldsymbol b_j^\mathsf{T}$ (use [np.outer](https://numpy.org/doc/stable/reference/generated/numpy.outer.html))
* just use numpy and calculate `A @ B`

The plan (see aslo dot product demo below):

* implement these four functions
* test that they return the same result
* measure their performance on a square matrix of shape $100\times 100$
* plot graphs of the execution time versus $n$

```python
import numpy as np

# 1. Nested Pythonic loop for matrix multiplication
def mat_mul_loop(A, B: np.array) -> np.array:
    m, n = A.shape
    n2, p = B.shape
    assert n == n2, "Matrix dimensions do not match for multiplication!"
    C = np.zeros((m, p))  # Initialize result matrix
    for i in range(m):
        for j in range(p):
            C[i, j] = sum(A[i, k] * B[k, j] for k in range(n))
    return C

# 2. Replace inner loop with np.dot
def mat_mul_inner(A, B: np.array) -> np.array:
    return np.dot(A, B)

# 3. Use np.outer for matrix multiplication
def mat_mul_outer(A, B: np.array) -> np.array:
    m, n = A.shape
    n2, p = B.shape
    assert n == n2, "Matrix dimensions do not match for multiplication!"
    C = np.zeros((m, p))  # Initialize result matrix
    for i in range(m):
        C[i, :] = np.sum(np.outer(A[i, :], B[:, :]), axis=0)
    return C

# 4. Use NumPy's built-in matrix multiplication
def mat_mul_np(A, B: np.array) -> np.array:
    return A @ B  # or np.matmul(A, B)
```

```python
# TESTING AREA

def TestMatrices(n_max):
    for n in range(1, n_max):
        m = np.random.randint(1, n_max)
        p = np.random.randint(1, n_max)
        A = np.random.randn(m, n)
        B = np.random.randn(n, p)
        prod_loop = mat_mul_loop(A, B)
        prod_inner = mat_mul_inner(A, B)
        prod_outer = mat_mul_outer(A, B)
        prod_np = mat_mul_np(A, B)
        assert np.allclose(prod_loop, prod_inner)
        assert np.allclose(prod_loop, prod_np)
        assert np.allclose(prod_loop, prod_outer)
```

## Demo for task 1.6: dot product performance

Compare several implementations of the dot product.

```python
import numpy as np

def dot_loop(a, b):
    result = 0
    for i in range(len(a)):
        result += a[i] * b[i]
    return result

def dot_sum(a, b):
    return sum(a * b)

def dot_np(a, b):
    return a @ b
```

Test that all this functions produce the same result:

```python
for n in range(1, 101):
    a = np.random.randn(n)
    b = np.random.randn(n)
    assert np.allclose(dot_loop(a, b), dot_sum(a, b))
    assert np.allclose(dot_loop(a, b), dot_np(a, b))
    assert np.allclose(dot_np(a, b), dot_sum(a, b))
```
